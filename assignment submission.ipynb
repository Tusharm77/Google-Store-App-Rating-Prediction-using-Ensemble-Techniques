```python
# Converting size to float 
df['Size'] = df['Size'].apply(lambda a : float(a))
```

```python
# Drop columns that you feel can not be used for model building. Example-
# App, Content Rating, Genre, Last updated, Current Ver, and Android Ver
# columns from the final data frame. (2 points)

df1 = df.drop(['App','Content Rating','Last Updated','Current Ver','Android Ver','Genres','Rating'], axis=1)
df1
```

```python
# Encode categorical column (Type, Rating_categories, Category) [ Hint - use get_dummies] (2 points)
df1.info()
```

```python
df2 = pd.get_dummies(df1,columns=['Type','Category'])
```

```python
# Since we are trying to predict the ratings category let's encode high rating as 2 and low rating as 1
df2['Rating_cat'] = df2.Rating_cat.str.replace('High','2')
df2['Rating_cat'] = df2.Rating_cat.str.replace('Low','1')
```

```python
df2.head(50)
```

```python
# Now our data is ready for bulding model to predict rating of a given app
```

```python
# Segregate dependent variable and independent features into two separate 
# variables and split the data into train and test set [ Use 70:30 split ]

from sklearn.model_selection import train_test_split

y = df2['Rating_cat']
X = df2.loc[:, df2.columns != 'Rating_cat']

X_train, X_test, y_train, y_test = train_test_split (X, y, test_size=0.3, random_state=42,)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# Building a Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier

dTree = DecisionTreeClassifier(criterion = 'gini', random_state=1)
dTree.fit(X_train, y_train)
```

```python
# Scoring our Decision Tree

print(dTree.score(X_train, y_train))
print(dTree.score(X_test, y_test))
```

```python
# Reducing over-fitting/ REgularization
dTreeR = DecisionTreeClassifier(criterion = 'gini', max_depth = 3, random_state=1)
dTreeR.fit(X_train, y_train)
print(dTreeR.score(X_train, y_train))
print(dTreeR.score(X_test, y_test))
```

```python
# Here, we can see that by Regularization and keeping the model simple, though the train data accuracy decreased
# the test data accuracy improve by around 5 %
```

```python
# Calculating the feature importance
imp1 = pd.DataFrame(dTreeR.feature_importances_, columns = ["Imp"], index = X_train.columns)
imp1 = imp1.sort_values(by=['Imp'],ascending=False)
imp1.head
```

```python
# Creating the confusion matrix
from sklearn.metrics import confusion_matrix
import seaborn as sns

print(dTreeR.score(X_test , y_test))
y_predict = dTreeR.predict(X_test)

cm = confusion_matrix(y_test, y_predict, labels=['1','2'])

df_cm = pd.DataFrame(cm, index = [i for i in ["1","2"]],
                  columns = [i for i in ["1","2"]])
plt.figure(figsize = (7,5))
sns.heatmap(df_cm, annot=True ,fmt='g')
```

# Building a Ensemble Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rfcl = RandomForestClassifier(n_estimators = 50, random_state=1,max_features=10)
rfcl = rfcl.fit(X_train, y_train)
```

```python
y_predict = rfcl.predict(X_test)
print(rfcl.score(X_test, y_test))
```

```python
# Creating confusion matrix
cm=confusion_matrix(y_test, y_predict,labels=['1','2'])

df_cm = pd.DataFrame(cm, index = [i for i in ["1","2"]],
                  columns = [i for i in ["1","2"]])
plt.figure(figsize = (7,5))
sns.heatmap(df_cm, annot=True ,fmt='g')
```

```python
# Feature importance of variables
# View a list of the features and their importance scores
imp2 = pd.DataFrame(rfcl.feature_importances_, columns = ["Imp"], index = X_train.columns)
imp2 = imp2.sort_values(by=['Imp'],ascending=False)
imp2
```

```python
print(rfcl.score(X_train, y_train))
print(rfcl.score(X_test, y_test))
```

# Gradient Boosting Model

```python
from sklearn.ensemble import GradientBoostingClassifier
gbcl = GradientBoostingClassifier(n_estimators = 50,random_state=1)
gbcl = gbcl.fit(X_train, y_train)
```

```python
y_predict = gbcl.predict(X_test)
print(gbcl.score(X_test, y_test))
cm=confusion_matrix(y_test, y_predict,labels=['1', '2'])

df_cm = pd.DataFrame(cm, index = [i for i in ["1","2"]],
                  columns = [i for i in ["1","2"]])
plt.figure(figsize = (7,5))
sns.heatmap(df_cm, annot=True ,fmt='g')
```

```python
# Feature importance of variables
# View a list of the features and their importance scores
imp3 = pd.DataFrame(gbcl.feature_importances_, columns = ["Imp"], index = X_train.columns)
imp3 = imp3.sort_values(by=['Imp'],ascending=False)
imp3
```

```python
print(gbcl.score(X_train, y_train))
print(gbcl.score(X_test, y_test))
```

# Creating a Stacked Model

```python
#The get_stacking() function below defines the StackingClassifier model by first defining a list of tuples 
#for the five base models, then defining the logistic regression meta-model to combine the predictions from 
#the base models using 5-fold cross-validation.

from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
```

```python
estimators = [('DT', dTreeR),('rf',rfcl),('GB',gbcl)]
clf = StackingClassifier(estimators=estimators, final_estimator=LogisticRegression())
clf.fit(X_train, y_train)
```

```python
y_predict = clf.predict(X_test)
print(clf.score(X_test, y_test))
cm=confusion_matrix(y_test, y_predict,labels=['1', '2'])

df_cm = pd.DataFrame(cm, index = [i for i in ["1","2"]],
                  columns = [i for i in ["1","2"]])
plt.figure(figsize = (7,5))
sns.heatmap(df_cm, annot=True ,fmt='g')
```

```python
print(clf.score(X_train, y_train))
print(clf.score(X_test, y_test))
```

# Stratified cross-validation of the models

```python
k = 10
from sklearn.model_selection  import StratifiedKFold, cross_val_score
stratified_kfold = StratifiedKFold(n_splits = k, random_state = 55)
result1 = cross_val_score(dTreeR, X, y, cv = stratified_kfold)
result2 = cross_val_score(rfcl, X, y, cv = stratified_kfold)
result3 = cross_val_score(gbcl, X, y, cv = stratified_kfold)
result4 = cross_val_score(clf, X, y, cv = stratified_kfold)
```

```python
result1
```

```python
data = [['Decision Tree',result1.mean(),result1.std()],['Random Forest',result2.mean(),result2.std()],['Gradient Boosting',result3.mean(),result3.std()],['Stacked Model',result4.mean(),result4.std()]]
Results = pd.DataFrame(data, columns=['Model','Mean','Std Dev'])
```
